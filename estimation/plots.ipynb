```python
import matplotlib.pyplot as plt
def three_plots(results_df, results_df_edge, results_df_2_latents, disitrbution_name):

    # Group the results by sample size and calculate the mean
    median_results = results_df.groupby('Sample Size').median()
    quantile_results_lower = results_df.groupby('Sample Size').quantile(0.25)
    quantile_results_upper = results_df.groupby('Sample Size').quantile(0.75)

    # Group the results by sample size and calculate the mean
    median_results_edge = results_df_edge.groupby('Sample Size').median()
    quantile_results_edge_lower = results_df_edge.groupby('Sample Size').quantile(0.25)
    quantile_results_edge_upper = results_df_edge.groupby('Sample Size').quantile(0.75)


    # Group the results by sample size and calculate the mean
    median_results_2_latents = results_df_2_latents.groupby('Sample Size').median()
    quantile_results_2_latents_lower = results_df_2_latents.groupby('Sample Size').quantile(0.25)
    quantile_results_2_latents_upper = results_df_2_latents.groupby('Sample Size').quantile(0.75)
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15, 5))

    plt.subplots_adjust(wspace=0.03)
    LEGEND_SIZE = 14
    SIZE = 18

    plt.rc('font', size=SIZE)
    plt.rc('axes', titlesize=SIZE)
    plt.rc('axes', labelsize=SIZE)
    plt.rc('xtick', labelsize=SIZE)
    plt.rc('ytick', labelsize=SIZE)
    plt.rc('legend', fontsize=LEGEND_SIZE)
    plt.rc('figure', titlesize=SIZE)

    # Plot for results_df_2_latents
    ax1.set_yscale('log')
    ax1.set_xscale('log')
    ax1.plot(median_results.index, median_results["Cumulant Error"], label='Cumulant', color="blue")
    ax1.fill_between(median_results.index, quantile_results_lower["Cumulant Error"], quantile_results_upper["Cumulant Error"], alpha=0.2, color="blue")
    ax1.plot(median_results.index, median_results["Cross Moment Error"], label='Cross Moment', color="red")
    ax1.fill_between(median_results.index, quantile_results_lower["Cross Moment Error"], quantile_results_upper["Cross Moment Error"], alpha=0.2, color="red")
    ax1.plot(median_results.index, median_results["GRICA Error"], label='GRICA', color="green")
    ax1.fill_between(median_results.index, quantile_results_lower["GRICA Error"], quantile_results_upper["GRICA Error"], alpha=0.2, color="green")
    ax1.plot(median_results.index, median_results["Relvlingam Error"], label='ReLVLiNGAM', color="purple")
    ax1.fill_between(median_results.index, quantile_results_lower["Relvlingam Error"], quantile_results_upper["Relvlingam Error"], alpha=0.2, color="purple")
    ax1.set_ylabel('Error (LOG-scale)')
    ax1.set_title(f"Graph: $\mathcal{{G}}_1$")
    ax1.grid()

    # Plot for results_df_2_latents

    ax2.set_yscale('log')
    ax2.set_xscale('log')
    ax2.plot(median_results_2_latents.index, median_results_2_latents["Cumulant Error"], label='Cumulant', color="blue")
    ax2.fill_between(median_results_2_latents.index, quantile_results_2_latents_lower["Cumulant Error"], quantile_results_2_latents_upper["Cumulant Error"], alpha=0.2, color="blue")
    ax2.plot(median_results_2_latents.index, median_results_2_latents["Cross Moment Error"], label='Cross Moment', color="red")
    ax2.fill_between(median_results_2_latents.index, quantile_results_2_latents_lower["Cross Moment Error"], quantile_results_2_latents_upper["Cross Moment Error"], alpha=0.2, color="red")
    ax2.plot(median_results_2_latents.index, median_results_2_latents["GRICA Error"], label='GRICA', color="green")
    ax2.fill_between(median_results_2_latents.index, quantile_results_2_latents_lower["GRICA Error"], quantile_results_2_latents_upper["GRICA Error"], alpha=0.2, color="green")
    ax2.plot(median_results_2_latents.index, median_results_2_latents["Relvlingam Error"], label='ReLVLiNGAM', color="purple")
    ax2.fill_between(median_results_2_latents.index, quantile_results_2_latents_lower["Relvlingam Error"], quantile_results_2_latents_upper["Relvlingam Error"], alpha=0.2, color="purple")
    ax2.set_title(f"Graph: $\mathcal{{G}}_2$")
    ax2.set_xlabel('Sample Size (LOG-scale)')
    ax2.grid()

    # Plot for results_df_edge
    ax3.set_yscale('log')
    ax3.set_xscale('log')
    ax3.plot(median_results_edge.index, median_results_edge["Cumulant Error"], label='Cumulant', color="blue")
    ax3.fill_between(median_results_edge.index, quantile_results_edge_lower["Cumulant Error"], quantile_results_edge_upper["Cumulant Error"], alpha=0.2, color="blue")
    ax3.plot(median_results_edge.index, median_results_edge["Cumulant Minimization Error"], label='Cumulant with Minimization', color="black")
    ax3.fill_between(median_results_edge.index, quantile_results_edge_lower["Cumulant Minimization Error"], quantile_results_edge_upper["Cumulant Minimization Error"], alpha=0.2, color="black")
    ax3.plot(median_results_edge.index, median_results_edge["Cross Moment Error"], label='Cross Moment', color="red")
    ax3.fill_between(median_results_edge.index, quantile_results_edge_lower["Cross Moment Error"], quantile_results_edge_upper["Cross Moment Error"], alpha=0.2, color="red")
    ax3.plot(median_results_edge.index, median_results_edge["GRICA Error"], label='GRICA', color="green")
    ax3.fill_between(median_results_edge.index, quantile_results_edge_lower["GRICA Error"], quantile_results_edge_upper["GRICA Error"], alpha=0.2, color="green")
    ax3.plot(median_results_edge.index, median_results_edge["Relvlingam Error"], label='ReLVLiNGAM', color="purple")
    ax3.fill_between(median_results_edge.index, quantile_results_edge_lower["Relvlingam Error"], quantile_results_edge_upper["Relvlingam Error"], alpha=0.2, color="purple")
    ax3.set_title(f"Graph: $\mathcal{{G}}_3$")
    ax3.legend(loc="lower left")
    ax3.grid()

    # Remove legends from ax1 and ax3
    ax1.legend().remove()
    ax3.legend().remove()

    # Add a general title
    fig.suptitle(f'Distribution: {disitrbution_name.capitalize()}', fontsize=SIZE +2)
    fig.subplots_adjust(top=.84)

    for ax in [ax1, ax2, ax3]:
        ax.set_xticks(median_results.index)

    # Add legend to ax2 and place it on the right of the figure
    ax3.legend(loc='lower center', bbox_to_anchor=(-0.64, -0.32), ncol=5)

    plt.savefig(f"Figures/{disitrbution_name}_G1_G2_G3_samples_vs_err.png", bbox_inches="tight")
    plt.show()

def g4_plots(df_gamma, df_beta):

    # Group the results by sample size and calculate the mean
    median_results_gamma = df_gamma.groupby('Sample Size').median()
    quantile_results_lower_gamma = df_gamma.groupby('Sample Size').quantile(0.25)
    quantile_results_upper_gamma = df_gamma.groupby('Sample Size').quantile(0.75)

    median_results_beta = df_beta.groupby('Sample Size').median()
    quantile_results_lower_beta = df_beta.groupby('Sample Size').quantile(0.25)
    quantile_results_upper_beta = df_beta.groupby('Sample Size').quantile(0.75)

    fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15, 5))

    plt.subplots_adjust(wspace=0.03)
    LEGEND_SIZE = 14
    SIZE = 16

    plt.rc('font', size=SIZE)
    plt.rc('axes', titlesize=SIZE)
    plt.rc('axes', labelsize=SIZE)
    plt.rc('xtick', labelsize=SIZE)
    plt.rc('ytick', labelsize=SIZE)
    plt.rc('legend', fontsize=LEGEND_SIZE)
    plt.rc('figure', titlesize=SIZE)

    # Plot for results_df_2_latents
    ax1.set_yscale('log')
    ax1.set_xscale('log')
    ax1.plot(median_results_gamma.index, median_results_gamma["Cumulant Error"], label='Cumulant', color="blue")
    ax1.fill_between(median_results_gamma.index, quantile_results_lower_gamma["Cumulant Error"], quantile_results_upper_gamma["Cumulant Error"], alpha=0.2, color="blue")
    ax1.plot(median_results_gamma.index, median_results_gamma["Cross Moment Error"], label='Cross Moment', color="red")
    ax1.fill_between(median_results_gamma.index, quantile_results_lower_gamma["Cross Moment Error"], quantile_results_upper_gamma["Cross Moment Error"], alpha=0.2, color="red")
    ax1.plot(median_results_gamma.index, median_results_gamma["GRICA Error"], label='GRICA', color="green")
    ax1.fill_between(median_results_gamma.index, quantile_results_lower_gamma["GRICA Error"], quantile_results_upper_gamma["GRICA Error"], alpha=0.2, color="green")
    ax1.plot(median_results_gamma.index, median_results_gamma["Relvlingam Error"], label='ReLVLiNGAM', color="purple")
    ax1.fill_between(median_results_gamma.index, quantile_results_lower_gamma["Relvlingam Error"], quantile_results_upper_gamma["Relvlingam Error"], alpha=0.2, color="purple")
    ax1.set_ylabel('Error (LOG-scale)')
    ax1.set_title(f"Disribution: Gamma")
    ax1.grid()
    # Plot for results_df_2_latents

    ax2.set_yscale('log')
    ax2.set_xscale('log')
    ax2.plot(median_results_beta.index, median_results_beta["Cumulant Error"], label='Cumulant', color="blue")
    ax2.fill_between(median_results_beta.index, quantile_results_lower_beta["Cumulant Error"], quantile_results_upper_beta["Cumulant Error"], alpha=0.2, color="blue")
    ax2.plot(median_results_beta.index, median_results_beta["Cross Moment Error"], label='Cross Moment', color="red")
    ax2.fill_between(median_results_beta.index, quantile_results_lower_beta["Cross Moment Error"], quantile_results_upper_beta["Cross Moment Error"], alpha=0.2, color="red")
    ax2.plot(median_results_beta.index, median_results_beta["GRICA Error"], label='GRICA', color="green")
    ax2.fill_between(median_results_beta.index, quantile_results_lower_beta["GRICA Error"], quantile_results_upper_beta["GRICA Error"], alpha=0.2, color="green")
    ax2.plot(median_results_beta.index, median_results_beta["Relvlingam Error"], label='ReLVLiNGAM', color="purple")
    ax2.fill_between(median_results_beta.index, quantile_results_lower_beta["Relvlingam Error"], quantile_results_upper_beta["Relvlingam Error"], alpha=0.2, color="purple")


    ax2.set_title(f"Disribution: Beta")
    # Remove x-axis labels from individual plots

    # Add a single x-axis label in the middle
    ax2.grid()

    # Remove legends from ax1 and ax3
    ax1.legend().remove()

    # Add a general title
    fig.suptitle(f'Graph: $\mathcal{{G}}_4$', fontsize=SIZE +2)
    fig.subplots_adjust(top=.84)
    fig.text(0.5, -0.01, 'Sample Size (LOG-scale)', ha='center', fontsize=SIZE)

    for ax in [ax1, ax2]:
        ax.set_xticks(median_results_gamma.index)

    # Add legend to ax2 and place it on the right of the figure
    ax2.legend(loc='lower center', bbox_to_anchor=(-0.1, -0.32), ncol=5)
    # Plot the results

    plt.savefig(f"Figures/G4_samples_vs_err.png", bbox_inches="tight")
    plt.show()
```

```python
import matplotlib.pyplot as plt
def iv_plots(df, distribution_name):

    # Group the results by sample size and calculate the mean
    median_results = df.groupby('Sample Size').median()
    quantile_results_lower = df.groupby('Sample Size').quantile(0.25)
    quantile_results_upper = df.groupby('Sample Size').quantile(0.75)

    MEDIUM_SIZE = 14
    BIGGER_SIZE = 16

    plt.rc('font', size=BIGGER_SIZE)
    plt.rc('axes', titlesize=BIGGER_SIZE)
    plt.rc('axes', labelsize=BIGGER_SIZE)
    plt.rc('xtick', labelsize=BIGGER_SIZE)
    plt.rc('ytick', labelsize=BIGGER_SIZE)
    plt.rc('legend', fontsize=MEDIUM_SIZE)
    plt.rc('figure', titlesize=BIGGER_SIZE)

    # Plot the results

    # Plot the mean results
    plt.yscale('log')
    plt.xscale('log')
    plt.plot(median_results.index, median_results["Cumulant Error"], label='Cumulant', color = "blue")

    plt.fill_between(median_results.index, quantile_results_lower["Cumulant Error"], quantile_results_upper["Cumulant Error"], alpha=0.2, color = "blue")

    plt.plot(median_results.index, median_results["Min Norm Error"],label='Min Norm',color = "red")

    plt.fill_between(median_results.index, quantile_results_lower["Min Norm Error"], quantile_results_upper["Min Norm Error"], alpha=0.2, color = "red")

    plt.plot(median_results.index, median_results["GRICA Error"],label='GRICA',color = "green")

    plt.fill_between(median_results.index, quantile_results_lower["GRICA Error"], quantile_results_upper["GRICA Error"], alpha=0.2, color = "green")

    plt.plot(median_results.index, median_results["Relvlingam Error"], label='ReLVLiNGAM', color="purple")
    plt.fill_between(median_results.index, quantile_results_lower["Relvlingam Error"], quantile_results_upper["Relvlingam Error"], alpha=0.2, color="purple")

    plt.xlabel('Sample Size (LOG-scale)')
    plt.ylabel('Error (LOG-scale)')
    plt.title(f"Distribution: {distribution_name.capitalize()}, Graph: IV")
    plt.legend()
    plt.legend(loc="lower left")
    plt.grid()
    plt.savefig(f"Figures/{distribution_name}_iv_samples_vs_err.png", bbox_inches="tight")
    plt.show()
```
